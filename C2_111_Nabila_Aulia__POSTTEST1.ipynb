## Algoritma Naive Bayes dan Preprocessing (pengenalan)

```python
# sebelum membuat algoritma pertama tama buatlah dataset baru dengan ketentuan
# mengambil dataset df

df = pd.read_csv("ChurnBank.csv")
```

```python
df
```

```python
# Menghapus kolom Unnamed: 0, CLIENTNUM, Umur, dan Limit Kartu Kredit (initinya yang berupa primary key dan bukan tipe kategori)
# dataset dengan kolom yang sudah dihapus tadi kemudian disimpan di dataframe baru (df baru)

data = df.drop(["Unnamed: 0", "CLIENTNUM", "Customer_Age", "Credit_Limit"], axis=1)
dfbaru = pd.DataFrame(data)
dfbaru
```

```python
# Buatlah sebuah fungsi naive bayes dengan ketentuan
# Mengambil data dari dataframe yang telah diproses pada blok sebelumnya (diatas)
# Nama fungsinya adalah naive_bayes
# Mengambil 1 saja parameter yaitu sebuah list berisi kategori

# Outputnya berupa prediksi Attrited Customer atau Existing Customer
# Outputnya berupa tuple (_hasilnya, persentasenya berapa) contoh (Existing Customer, 0.004)

# Tidak boleh menggunakan library manapun selain pandas dan numpy


# contoh
# def naive_bayes(prediksi:list)-> tuple :
#   ........

# jadi nanti saat saya coba masukan data 
# naive_bayes(["F",	"Uneducated","Single", "Unknown	Blue"]) akan keluar tuple berupa ("Attired Customer", 0.002)

set_Attrited = dfbaru[dfbaru["Attrition_Flag"] == "Attrited Customer"].shape[0]
set_Existing = dfbaru[dfbaru["Attrition_Flag"] == "Existing Customer"].shape[0]
count = dfbaru.shape[0]

prob_Attrited = set_Attrited / count
prob_Existing = set_Existing / count 

categories = ["Gender", "Education_Level", "Marital_Status", "Card_Category"]

p_categories = {}
for col in categories:
    p_categories[col] = {}
    for category in dfbaru[col].unique():
      Attrited_cat = dfbaru[(dfbaru["Attrition_Flag"] == "Attrited Customer") & (dfbaru[col] == category)].shape[0]
      Existing_cat = dfbaru[(dfbaru["Attrition_Flag"] == "Attrited Customer") & (dfbaru[col] == category)].shape[0]
      pro_Att = Attrited_cat / set_Attrited
      pro_Exist = Existing_cat / set_Existing
      p_categories[col][category] = {"Attrited Customer": pro_Att, "Existing Customer": pro_Exist}

def naive_bayes(prediksi:list) -> tuple:
  pre_Attrited = set_Attrited
  pre_Existing = set_Existing
  for i, col in enumerate(categories):
    category = prediksi[i]
    pre_Attrited *= p_categories[col][category]["Attrited Customer"]
    pre_Existing *= p_categories[col][category]["Existing Customer"]

  if pre_Attrited < pre_Existing:
    hasil = "Attrited Customer"
    persentase = pre_Attrited / (pre_Attrited + pre_Existing)
  else:
    hasil = "Existing Customer"
    persentase = pre_Existing / (pre_Attrited + pre_Existing)

  return hasil, persentase


print("Format input data yaitu F, Uneducated, Single, Blue < harus menggunakan (,) tanda koma dan spasi. Jika tidak akan terjadi error >")

data = input("\nMasukkan data: ")
data_list = data.split(", ")

# memanggil fungsi naive_bayes
hasil, persentase = naive_bayes(data_list)

# output
print("hasil output: ", (hasil, persentase))
```
